```python
base_url = "https://www.udemy.com/api-2.0/discovery-units/all_courses/?p={0}&page_size=60&subcategory=&instructional_level=&lang=&price=&duration=&closed_captions=&subs_filter_type=&category_id={1}&source_page=category_page&locale=ko_KR&currency=krw&navigation_locale=en_US&skip_price=true&sos=pc&fl=cat"

categories = {
    288: "development",
    268: "business",
    294: "IT_and_Software",
    292: "business_productivity",
    296: "self_improvement",
    269: "design",
    290: "marketing",
    274: "lifestyle",
    273: "pictures_and_videos",
    276: "health_and_fitness",
    278: "music",
    300: "education"
    328: "finance-and-accounting"
}


all_courses_df = pd.DataFrame(columns=["id", "image_url", "instructional_level"])

for category_id, category_name in categories.items():
    print(f"Fetching data for {category_name} category...")
    for page in range(1, 626):
        url = base_url.format(page, category_id)
        response = requests.get(url)
        data = response.json()
        courses = data.get("unit", {}).get("items", [])
        print(f"Fetching data for {category_name} category... {page}")
        
        for course in courses:
            course_id = course.get("id")
            row = {
                "id": course_id,
                "image_url": course.get("image_125_H"),
                "instructional_level": course.get("instructional_level")
            }
            all_courses_df = all_courses_df.append(row, ignore_index=True)
    
    print(f"Data fetching complete for {category_name} category.")

    csv_filename = f"{category_name}_courses.csv"
    all_courses_df.to_csv(csv_filename, index=False)
    print(f"Data saved to {csv_filename}")

    
    all_courses_df = pd.DataFrame(columns=["id", "image_url", "instructional_level"])
```

```python
def merge_csv_files_by_category(categories):
    combined_df = pd.DataFrame()
    current_directory = os.getcwd() 
    for category_id, category_name in categories.items():
        category_file_name = f"{category_name}_courses.csv"
        
        print(f"Merging {category_name} category...")
        category_file_path = os.path.join(current_directory, category_file_name)
        if os.path.exists(category_file_path):
            df = pd.read_csv(category_file_path)
            combined_df = pd.concat([combined_df, df], ignore_index=True)
        print(f"Finish merging {category_name} category...")
    return combined_df

categories = {
    288: "development",
    268: "business",
    294: "IT_and_Software",
    292: "business_productivity",
    296: "self_improvement",
    269: "design",
    290: "marketing",
    274: "lifestyle",
    273: "pictures_and_videos",
    276: "health_and_fitness",
    278: "music",
    300: "education",
    328: "finance-and-accounting"
}

merged_df = merge_csv_files_by_category(categories)

merged_df.to_csv("merged_category.csv", index=False)
```
